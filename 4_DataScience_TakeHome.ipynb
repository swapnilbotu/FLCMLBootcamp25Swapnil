```python
import json
import os
import pandas as pd
import requests
from IPython.display import display # Import display
```

```python
# Fetch the JSON 'grocery inventory' data from the URL
url = 'https://raw.githubusercontent.com/shashankbl/mlbootcamp2025flc/main/extras/grocery_inventory.json'  # Use 'raw.githubusercontent.com'
try:
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for bad status codes
    data = response.json()
except requests.exceptions.RequestException as e:
    print(f"Error fetching JSON from URL: {e}")
    data = {}  # Initialize as empty dictionary in case of error

# Convert the JSON data to a Pandas DataFrame
try:
    inventory_df = pd.DataFrame(data)
    display(inventory_df) # Display the DataFrame using display() for pretty print
except (ValueError, KeyError) as e:
    print(f"Error converting JSON to DataFrame: {e}")
    if data:
        print("JSON data loaded but could not be formatted as a table")
```

```python
# Fetch the JSON 'transaction history' data from the URL
url = 'https://raw.githubusercontent.com/shashankbl/mlbootcamp2025flc/main/extras/transaction_history.json'  # Use 'raw.githubusercontent.com'
try:
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for bad status codes
    data = response.json()
except requests.exceptions.RequestException as e:
    print(f"Error fetching JSON from URL: {e}")
    data = {}  # Initialize as empty dictionary in case of error

# Convert the JSON data to a Pandas DataFrame
try:
    transaction_df = pd.DataFrame(data)
    display(transaction_df) # Display the DataFrame using display() for pretty print
except (ValueError, KeyError) as e:
    print(f"Error converting JSON to DataFrame: {e}")
    if data:
        print("JSON data loaded but could not be formatted as a table")
```

### Calculate the value of prior transactions.

```python
# Create a lookup for unit costs by item
unit_cost_dict = inventory_df.set_index('item')['unit_cost'].to_dict()

def compute_transaction_value(qty_dict):
    total = 0
    for item, qty in qty_dict.items():
        total += qty * unit_cost_dict.get(item, 0)
    return total

# Compute transaction value for each transaction
transaction_df['transaction_value'] = transaction_df['quantities'].apply(compute_transaction_value)

print("Total value of prior transactions: $", round(transaction_df['transaction_value'].sum(), 2))
```

### How many transactions violated the 'max quantities' rule?

```python
# Write code here...
# Create a lookup for max quantities by item
max_quantity_dict = inventory_df.set_index('item')['max_quantity'].to_dict()

def violates_max_rule(qty_dict):
    # Return True if any item's purchased quantity exceeds its max_quantity
    for item, qty in qty_dict.items():
        if qty > max_quantity_dict.get(item, float('inf')):
            return True
    return False

# Apply the function to flag transactions that violate the rule
transaction_df['violation'] = transaction_df['quantities'].apply(violates_max_rule)

# Count the transactions that violated the max quantities rule
violation_count = transaction_df['violation'].sum()
print("Number of transactions violating the max quantity rule:", violation_count)
```

### Track the popularity of items in the grocery store by factoring the transaction quantities (histogram representation)


```python
import matplotlib.pyplot as plt
from collections import defaultdict

# Sum total quantities sold per item across all transactions
item_sales = defaultdict(int)
for qty_dict in transaction_df['quantities']:
    for item, qty in qty_dict.items():
        item_sales[item] += qty

# Convert to a DataFrame/Series for plotting
import pandas as pd
item_sales_series = pd.Series(item_sales).sort_values()

plt.figure(figsize=(10,6))
item_sales_series.plot(kind='bar')
plt.xlabel("Item")
plt.ylabel("Total Quantity Sold")
plt.title("Popularity of Items in the Grocery Store")
plt.tight_layout()
plt.show()
```

### Average transaction per day (ignore days with no transaction as they were holidays)

```python

# Convert the date column from list to datetime
transaction_df['date'] = transaction_df['date'].apply(lambda d: pd.Timestamp(year=d[0], month=d[1], day=d[2]))

# Group by date and calculate total transaction value per day
daily_transaction_value = transaction_df.groupby('date')['transaction_value'].sum()

# Compute the average of daily totals (only days with transactions)
average_daily_value = daily_transaction_value.mean()
print("Average transaction revenue per day: $", round(average_daily_value, 2))
```

### Identify the three least efficient transactions and three most efficient transactions (based on revenue/min metric)

```python
# Write code here...
# Ensure the transaction_duration column is numeric
transaction_df['transaction_duration'] = pd.to_numeric(transaction_df['transaction_duration'], errors='coerce')

# Now compute efficiency as revenue per minute (using true division)
transaction_df['efficiency'] = transaction_df['transaction_value'] / transaction_df['transaction_duration']

# Identify the three most efficient transactions
most_efficient = transaction_df.nlargest(3, 'efficiency')

# Identify the three least efficient transactions
least_efficient = transaction_df.nsmallest(3, 'efficiency')

print("Three most efficient transactions (revenue per minute):")
display(most_efficient[['id', 'customer', 'transaction_value', 'transaction_duration', 'efficiency']])

print("Three least efficient transactions (revenue per minute):")
display(least_efficient[['id', 'customer', 'transaction_value', 'transaction_duration', 'efficiency']])
```

### Create a visualization of individual items sold over time

```python
# Write code here...
# Build a DataFrame with one row per (date, item) occurrence
sales_records = []
for idx, row in transaction_df.iterrows():
    trans_date = row['date']
    for item, qty in row['quantities'].items():
        sales_records.append({'date': trans_date, 'item': item, 'quantity': qty})

sales_df = pd.DataFrame(sales_records)

# Pivot the DataFrame so each item is a column and dates are the index
sales_over_time = sales_df.pivot_table(index='date', columns='item', values='quantity', aggfunc='sum', fill_value=0)

plt.figure(figsize=(12,6))
sales_over_time.plot(marker='o')
plt.xlabel("Date")
plt.ylabel("Quantity Sold")
plt.title("Individual Items Sold Over Time")
plt.legend(title="Item", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Based on the transaction history, calculate the total inventory at store opening (prior transaction quantities + current inventory quantities)

```python
# Write code here...
# First, compute the total quantity sold per item from the transactions
sold_df = pd.Series(item_sales, name='sold_quantity').reset_index().rename(columns={'index': 'item'})

#merge with inventary data (current inventory available)
opening_inventory_df = pd.merge(inventory_df, sold_df, on='item', how='left')
opening_inventory_df['sold_quantity'] = opening_inventory_df['sold_quantity'].fillna(0)

# Calculate store opening inventory per item (current inventory + sold quantities)
opening_inventory_df['opening_inventory'] = opening_inventory_df['quantity'] + opening_inventory_df['sold_quantity']

print("Store Opening Inventory by Item:")
display(opening_inventory_df[['item', 'opening_inventory']])
```
